```python
import matplotlib.pyplot as plt
import numpy as np

# Models (hourly)
models = [
    "ARIMA", "SARIMA", "Linear Regression", "XGBoost", 
    "LSTM", "Sentiment-Enhanced LSTM"
]

rmse = [
    0.0051836, 0.0051637, 0.0051637, 0.0019759,
    0.02916, 0.0074646
]

mape = [
    3.12, 3.10, 3.10, 2.46,
    88.27, 19.86
]

# Bar Plot for RMSE
plt.figure(figsize=(10,5))
plt.bar(models, rmse, color="skyblue")
plt.title("RMSE Comparison - Hourly Data Models")
plt.ylabel("RMSE")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()

# Bar Plot for Safe-MAPE
plt.figure(figsize=(10,5))
plt.bar(models, mape, color="salmon")
plt.title("MAPE Comparison - Hourly Data Models")
plt.ylabel("MAPE (%)")
plt.xticks(rotation=45)
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```
