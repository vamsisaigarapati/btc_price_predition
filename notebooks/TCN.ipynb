```python
import pandas as pd
import torch
from pytorch_forecasting import TimeSeriesDataSet
from pytorch_forecasting.data import GroupNormalizer

# Assume df is already loaded as you showed
df = df.copy()

# Create a running time index
df['time_idx'] = range(len(df))

# Add a group ID (required even if only 1 series)
df['group_id'] = 'bitcoin'

print(df.head())
```

```python
# Features and Target
X = df.drop(columns=["Close", "datetime", "group_id"])
y = df["Close"]

# Convert to torch tensors
X_tensor = torch.tensor(X.values, dtype=torch.float32)
y_tensor = torch.tensor(y.values, dtype=torch.float32)

print(X_tensor.shape, y_tensor.shape)
```

```python
from torch.utils.data import Dataset

class TimeSeriesDataset(Dataset):
    def __init__(self, X, y, input_window, output_window):
        self.X = X
        self.y = y
        self.input_window = input_window
        self.output_window = output_window

    def __len__(self):
        return len(self.X) - self.input_window - self.output_window

    def __getitem__(self, idx):
        x_seq = self.X[idx : idx + self.input_window]
        y_seq = self.y[idx + self.input_window : idx + self.input_window + self.output_window]
        return x_seq, y_seq
```

```python
input_window = 24
output_window = 6

# Train-Validation Split
split_idx = int(0.8 * len(X_tensor))
X_train, X_val = X_tensor[:split_idx], X_tensor[split_idx:]
y_train, y_val = y_tensor[:split_idx], y_tensor[split_idx:]

# Create Dataset
train_dataset = TimeSeriesDataset(X_train, y_train, input_window, output_window)
val_dataset = TimeSeriesDataset(X_val, y_val, input_window, output_window)

# Create DataLoaders
batch_size = 64
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
```

```python
import torch.nn as nn

class TCNModel(nn.Module):
    def __init__(self, input_size, output_size, num_channels=[32, 32], kernel_size=3, dropout=0.2):
        super(TCNModel, self).__init__()
        self.tcn = nn.Sequential(
            nn.Conv1d(input_size, num_channels[0], kernel_size, padding=kernel_size//2),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Conv1d(num_channels[0], num_channels[1], kernel_size, padding=kernel_size//2),
            nn.ReLU(),
            nn.Dropout(dropout)
        )
        self.linear = nn.Linear(num_channels[-1], output_size)

    def forward(self, x):
        # x: [batch_size, input_window, features]
        x = x.permute(0, 2, 1)  # [batch_size, features, input_window]
        y1 = self.tcn(x)         # [batch_size, channels, input_window]
        y1 = y1[:, :, -1]        # take last time step output
        out = self.linear(y1)
        return out
```

```python
# Setup Model
input_size = X_tensor.shape[1]   # number of features
output_size = output_window      # predict next 6 values

model = TCNModel(input_size=input_size, output_size=output_size)
model = model.to('cuda' if torch.cuda.is_available() else 'cpu')

# Setup Optimizer and Loss
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()
```

```python
# Train Model
epochs = 30

device = 'cuda' if torch.cuda.is_available() else 'cpu'

for epoch in range(epochs):
    model.train()
    train_losses = []

    for x_batch, y_batch in train_loader:
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)

        optimizer.zero_grad()
        output = model(x_batch)

        loss = loss_fn(output, y_batch)
        loss.backward()
        optimizer.step()

        train_losses.append(loss.item())

    val_losses = []
    model.eval()
    with torch.no_grad():
        for x_val, y_val in val_loader:
            x_val, y_val = x_val.to(device), y_val.to(device)
            val_output = model(x_val)
            val_loss = loss_fn(val_output, y_val)
            val_losses.append(val_loss.item())

    print(f"Epoch {epoch+1}: Train Loss = {sum(train_losses)/len(train_losses):.4f}, Validation Loss = {sum(val_losses)/len(val_losses):.4f}")
```
