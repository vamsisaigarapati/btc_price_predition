```python
import json
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Define file paths
parquet_file_path = "../datasets/normalised_bitcoin_price.parquet"
json_file_path = "../datasets/scaler_params.json"

# Ensure dataset directory exists
import os
os.makedirs("./datasets", exist_ok=True)

# Preserve the date column separately
btc_data.reset_index(inplace=True)  # Ensure 'date' is a regular column

# Selecting relevant columns (excluding 'date' for scaling)
features = ['Open', 'High', 'Low', 'Close', 'Volume']
btc_data_subset = btc_data[features]

# Initialize and apply MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
btc_data_scaled = pd.DataFrame(scaler.fit_transform(btc_data_subset), columns=features)

# Restore the 'date' column
btc_data_scaled.insert(0, 'date', btc_data['date'])

# Save scaled data as a Parquet file
btc_data_scaled.to_parquet(parquet_file_path, index=False)

# Save Min/Max values in JSON
scaler_params = {
    "min": btc_data_subset.min().to_dict(),
    "max": btc_data_subset.max().to_dict()
}
with open(json_file_path, "w") as f:
    json.dump(scaler_params, f, indent=4)

print(f"Normalized data saved at: {parquet_file_path}")
print(f"Scaler parameters saved at: {json_file_path}")
```

# Bitcoin tweets dataset cleaning

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("kaushiksuresh147/bitcoin-tweets")

print("Path to dataset files:", path)
```

```python
import os
files = os.listdir(path)

# Read the CSV files into separate DataFrames
csv_files = [file for file in files if file.endswith('.csv')]
bitcoin_tweets_1 = pd.read_csv(os.path.join(path, csv_files[0]))
# bitcoin_tweets_2 = pd.read_csv(os.path.join(path, csv_files[1]))

# bitcoin_tweets_1.head()
# bitcoin_tweets_2.head()
```

```python
# Number of records
num_records_1 = len(bitcoin_tweets_1)
print(f"Number of records in bitcoin_tweets_1: {num_records_1}")

# First few rows
bitcoin_tweets_1.head()

# Unique date values
unique_dates_1 = bitcoin_tweets_1['date'].unique()
print(unique_dates_1)
```

```python
bitcoin_tweets = bitcoin_tweets_1.copy()
```

```python
bitcoin_tweets.head()
```

```python
num_records = len(bitcoin_tweets)
print(f"Number of records in bitcoin_tweets: {num_records}")
```

```python
# Create a mask to identify rows that cannot be converted to datetime
invalid_date_mask = pd.to_datetime(bitcoin_tweets['date'], errors='coerce').isna()

# Filter the original DataFrame using the mask
invalid_date_records = bitcoin_tweets[invalid_date_mask].copy()

# Display the records with invalid date format
# invalid_date_records.head()
```

```python
# Identify the columns that need to be shifted
columns_to_shift = ['user_description', 'user_created', 'user_followers', 'user_friends', 'user_favourites', 'user_verified', 'date', 'text', 'hashtags', 'source', 'is_retweet']

# Shift the values two columns to the right for the invalid records
invalid_date_records[columns_to_shift] = invalid_date_records[columns_to_shift].shift(2, axis=1)

# Fill the first two columns with NaN values
invalid_date_records[['user_name', 'user_location']] = invalid_date_records[['user_description', 'user_created']]

# Update the original DataFrame with the corrected invalid records
bitcoin_tweets.update(invalid_date_records)

# Display the corrected invalid records
invalid_date_records.head()
```

```python
# Remove the faulty records from the original DataFrame
bitcoin_tweets = bitcoin_tweets[~invalid_date_mask]

# Attach the corrected records
bitcoin_tweets = pd.concat([bitcoin_tweets, invalid_date_records])

# Sort the values by date
bitcoin_tweets = bitcoin_tweets.sort_values(by='date')

# Display the first few rows of the sorted DataFrame
bitcoin_tweets.head()
```

```python
len(bitcoin_tweets)
```

```python
selected_columns = [
    'date',               # Timestamp of the tweet
    'text',               # Tweet content (for sentiment analysis)
    'hashtags',           # Hashtags used (trend analysis)
    'user_followers',     # Number of followers (influence factor)
    'user_verified',      # Boolean - Verified user or not
    'is_retweet',         # Boolean - Whether it's a retweet
    'source'              # Tweet source (e.g., Twitter Web App)
]
```

```python
bitcoin_tweets.dtypes
```

```python
bitcoin_tweets['user_created'] = pd.to_datetime(bitcoin_tweets['user_created'], errors='coerce')
bitcoin_tweets['date'] = pd.to_datetime(bitcoin_tweets['date'], errors='coerce')

# Convert numeric columns to integers
bitcoin_tweets['user_followers'] = pd.to_numeric(bitcoin_tweets['user_followers'], errors='coerce').fillna(0).astype('int64')
bitcoin_tweets['user_friends'] = pd.to_numeric(bitcoin_tweets['user_friends'], errors='coerce').fillna(0).astype('int64')
bitcoin_tweets['user_favourites'] = pd.to_numeric(bitcoin_tweets['user_favourites'], errors='coerce').fillna(0).astype('int64')

# Convert boolean columns
bitcoin_tweets['user_verified'] = bitcoin_tweets['user_verified'].astype('bool')
bitcoin_tweets['is_retweet'] = bitcoin_tweets['is_retweet'].astype('bool')

# Convert categorical columns
bitcoin_tweets['source'] = bitcoin_tweets['source'].astype('category')

# Save the cleaned dataset as a Parquet file
output_path = "bitcoin_tweets_cleaned.parquet"
bitcoin_tweets.to_parquet(output_path, engine='pyarrow', index=False)

print(f"✅ Cleaned dataset saved as {output_path}")
```

```python
import matplotlib.pyplot as plt

# Ensure the 'date' column is in datetime format
# bitcoin_tweets['date'] = pd.to_datetime(bitcoin_tweets['date'], errors='coerce')

# Group by date and count the number of tweets per day
tweets_per_day = bitcoin_tweets.groupby(bitcoin_tweets['date'].dt.date).size()

# Plot the number of tweets over time
plt.figure(figsize=(12, 6))
tweets_per_day.plot()
plt.title('Number of Bitcoin Tweets Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Tweets')
plt.grid(True)
plt.show()
```
